## **4. Constru√ß√£o do Dataset Final**
Realize 5 jun√ß√µes entre os DataFrames. Siga a ordem dos arquivos informados na etapa 2.

```python
# 1¬™ jun√ß√£o: results + races
df = results_df.merge(races_df, on="raceId", how="inner")

# 2¬™ jun√ß√£o: + circuits
df = df.merge(circuits_df, on="circuitId", how="inner")

# 3¬™ jun√ß√£o: + constructors
df = df.merge(constructors_df, on="constructorId", how="inner")

# 4¬™ jun√ß√£o: + drivers (corrigido com suffixes)
df = df.merge(drivers_df, on="driverId", how="inner", suffixes=("", "_driver"))

# 5¬™ jun√ß√£o: + qualifying (corrigido com suffixes)
df = df.merge(
    qualifying_df,
    on=["raceId", "driverId", "constructorId"],
    how="left",
    suffixes=("", "_qualifying")
)

print("Dataset Final Constru√≠do com 5 jun√ß√µes (corrigido)")
df.head()
```

## **7. Limpeza e Sele√ß√£o de Atributos**
Elimine colunas irrelevantes, que voc√™ julgar necess√°rias, com `df.drop()`.

```python
# Removendo as colunas irrelevantes
df = df.drop(columns=[
    "url", "url_driver", "url_qualifying",   # links
    "number", "number_driver",               # IDs redundantes
    "positionText", "positionOrder",         # vers√µes duplicadas de posi√ß√£o
    "time", "time_qualifying",               # v√°rias vezes duplicadas
    "milliseconds", "milliseconds_qualifying"
], errors="ignore")

print("\033[1mColunas ap√≥s limpeza:\033[0m")
print(df.columns.tolist())
```

## 8. Filtragem Temporal
Mantenha apenas os dados a partir de 2012.

```python
# Filtrando os dados a partir de 2012
df = df[df["year"] >= 2012]

print(f"\033[1mN√∫mero de registros ap√≥s filtro:\033[0m {len(df)}")
```

## 9. Renomea√ß√£o de Colunas e Convers√µes
Renomeie colunas (ex: name para `GP_name`), concatene nomes de pilotos (forename + surname),  e converta datas (dob e date)  com `pd.to_datetime()`.

```python
# Renomear colunas
df = df.rename(columns={
    "name": "GP_name",
    "constructorRef": "team",
    "surname": "driver_surname",
    "forename": "driver_forename",
    "dob": "driver_dob",
    "date": "race_date"
})

# Concatenar nome do piloto
df["driver_name"] = df["driver_forename"] + " " + df["driver_surname"]

# Converter datas
df["driver_dob"] = pd.to_datetime(df["driver_dob"], errors="coerce")
df["race_date"] = pd.to_datetime(df["race_date"], errors="coerce")

print("\033[1mColunas renomeadas e datas convertidas:\033[0m")
df[["driver_name", "driver_dob", "race_date", "GP_name", "team"]].head()
```

## **10. C√°lculo da Idade do Piloto**
Crie a coluna `age_at_gp_in_days` com a diferen√ßa entre date e dob.

```python
# Calcular idade do piloto em dias no GP
df["age_at_gp_in_days"] = (df["race_date"] - df["driver_dob"]).dt.days

# Visualizar
df[["driver_name", "race_date", "driver_dob", "age_at_gp_in_days"]].head()
```

## **11. Padroniza√ß√£o de Nomes de Construtores (Equipes)**
Use `df['constructor'].apply()` para realizar as seguintes substitui√ß√µes:
- 'Force India' ou 'Racing Point' ‚Üí 'Aston Martin'
- 'Sauber' ‚Üí 'Alfa Romeo'
- 'Lotus F1', 'Renault', 'Alpine F1 Team' ‚Üí 'Alpine F1'
- 'Toro Rosso' ‚Üí 'AlphaTauri'

```python
def normalize_constructor(name):
    if name in ["Force India", "Racing Point"]:
        return "Aston Martin"
    elif name == "Sauber":
        return "Alfa Romeo"
    elif name in ["Lotus F1", "Renault", "Alpine F1 Team"]:
        return "Alpine F1"
    elif name == "Toro Rosso":
        return "AlphaTauri"
    else:
        return name

# Criar coluna padronizada a partir de 'team'
df["constructor_clean"] = df["team"].apply(normalize_constructor)

# Conferir valores √∫nicos
df[["team", "constructor_clean"]].drop_duplicates().head(15)
```

## **12. Padroniza√ß√£o de Pa√≠ses e Nacionalidades**
Reduza os nomes de nacionalidades e pa√≠ses para 3 letras. Normalize com `replace` e `apply`.

```python
# Mapeamento de nacionalidades (exemplo, pode expandir conforme necess√°rio)
nationality_map = {
    "British": "GBR",
    "German": "GER",
    "Italian": "ITA",
    "Brazilian": "BRA",
    "French": "FRA",
    "Spanish": "ESP",
    "Dutch": "NED",
    "Finnish": "FIN",
    "Mexican": "MEX",
    "Canadian": "CAN",
    "Australian": "AUS",
    "American": "USA",
    "Russian": "RUS",
    "Japanese": "JPN"
}

# Normalizar nacionalidade
df["driver_nat"] = df["nationality"].replace(nationality_map)

# Normalizar pa√≠ses (se existir colunas 'country' ou 'location')
if "country" in df.columns:
    country_map = {
        "UK": "GBR",
        "United Kingdom": "GBR",
        "Germany": "GER",
        "Italy": "ITA",
        "Brazil": "BRA",
        "France": "FRA",
        "Spain": "ESP",
        "Netherlands": "NED",
        "Finland": "FIN",
        "Mexico": "MEX",
        "Canada": "CAN",
        "Australia": "AUS",
        "USA": "USA",
        "United States": "USA",
        "Russia": "RUS",
        "Japan": "JPN"
    }
    df["country_clean"] = df["country"].replace(country_map)

df[["driver_name", "driver_nat"]].head(10)
```

## *13. Indicadores de Corrida em Casa*
Crie colunas driver_home e constructor_home com compara√ß√£o entre nacionalidade e pa√≠s (exemplo: `driver_nationality == country` ).

```python
# Indicadores de corrida em casa (piloto)
df["driver_home"] = np.where(df["driver_nat"] == df["country_clean"], 1, 0)

# Indicadores de corrida em casa (equipe)
# (usando constructor_clean nacionalidade -> simplifica√ß√£o, pois pode precisar mapear equipe->pa√≠s)
team_country_map = {
    "Ferrari": "ITA",
    "Mercedes": "GER",
    "Red Bull": "AUT",
    "McLaren": "GBR",
    "Alpine F1": "FRA",
    "AlphaTauri": "ITA",
    "Aston Martin": "GBR",
    "Williams": "GBR",
    "Alfa Romeo": "SUI",
    "Haas F1 Team": "USA"
}
df["constructor_country"] = df["constructor_clean"].map(team_country_map)
df["constructor_home"] = np.where(df["constructor_country"] == df["country_clean"], 1, 0)

df[["driver_name", "driver_nat", "country_clean", "driver_home", "constructor_clean", "constructor_home"]].head(10)
```

## **14. Indicadores de Abandono (DNF)**
Use `statusId` para gerar `driver_dnf` e `constructor_dnf`.

```python
# Indicador de abandono (piloto)
df["driver_dnf"] = np.where(df["statusId"] == 1, 0, 1)

# Indicador de abandono por equipe: se algum piloto da equipe abandonou, conta como DNF
df["constructor_dnf"] = df.groupby(["raceId", "constructor_clean"])["driver_dnf"].transform("max")

df[["driver_name", "constructor_clean", "statusId", "driver_dnf", "constructor_dnf"]].head(10)
```

## **15. An√°lise Explorat√≥ria com Seaborn e Matplotlib**
Use seaborn e matplotlib para criar gr√°ficos como histograma, boxplot, barra e dispers√£o.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar coluna auxiliar: idade em anos
df["age_years"] = df["age_at_gp_in_days"] // 365

# -----------------------------
# Histograma: idade dos pilotos
# -----------------------------
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="age_years", bins=20, kde=True)
plt.title("Distribui√ß√£o da Idade dos Pilotos (em anos)")
plt.xlabel("Idade (anos)")
plt.ylabel("Frequ√™ncia")
plt.show()

# -----------------------------
# Boxplot: posi√ß√µes finais por equipe
# -----------------------------
# Usar 'position' e 'name_y' (equipe)
df_box = df[['name_y','position']].dropna()

plt.figure(figsize=(12,6))
sns.boxplot(data=df_box, x="name_y", y="position", showfliers=False)
plt.xticks(rotation=45)
plt.title("Distribui√ß√£o de Posi√ß√µes por Equipe")
plt.xlabel("Equipe")
plt.ylabel("Posi√ß√£o Final")
plt.show()

# -----------------------------
# Gr√°fico de barras: vit√≥rias por piloto
# -----------------------------
winners = df[df["position"] == 1]['name_x'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=winners.values, y=winners.index, palette="viridis")
plt.title("Top 10 Pilotos em Vit√≥rias (2012+)")
plt.xlabel("N√∫mero de Vit√≥rias")
plt.ylabel("Piloto")
plt.show()

# -----------------------------
# Dispers√£o: idade do piloto vs posi√ß√£o final
# -----------------------------
df_scatter = df[['age_years','position','name_y']].dropna()

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df_scatter,
    x="age_years",
    y="position",
    hue="name_y",
    alpha=0.6
)
plt.title("Idade do Piloto vs Posi√ß√£o Final")
plt.xlabel("Idade (anos)")
plt.ylabel("Posi√ß√£o Final")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title="Equipe")
plt.show()
```

## **16. Confiabilidade dos Pilotos**
Calcule a confiabilidade com base na propor√ß√£o de abandonos.

Visualize com gr√°ficos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Criar coluna bin√°ria: 1 = abandono, 0 = terminou
df['abandoned'] = df['position'].isna().astype(int)

# Calcular confiabilidade por piloto (usar 'name_x' como piloto)
reliability = df.groupby('name_x')['abandoned'].agg(['sum','count']).reset_index()
reliability['reliability'] = 1 - (reliability['sum'] / reliability['count'])

# Top 10 pilotos mais confi√°veis
top_reliable = reliability.sort_values(by='reliability', ascending=False).head(10)

# Gr√°fico
plt.figure(figsize=(10,6))
sns.barplot(x='reliability', y='name_x', data=top_reliable, palette='magma')
plt.title('Top 10 Pilotos Mais Confi√°veis (Menos Abandonos)')
plt.xlabel('Confiabilidade')
plt.ylabel('Piloto')
plt.xlim(0,1)
plt.show()
```
